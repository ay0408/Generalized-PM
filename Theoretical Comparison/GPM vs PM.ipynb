```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def V0(b):
    eps = 2; x = math.exp(eps/2)
    z = np.zeros(len(b))
    for i in range(len(b)):
        y = math.exp(eps/b[i])
        z[i] = ((y+1)*(x**4 + 3*(x**2)*y + 3*(y**2) + y**3))/(3*((x**2-1)**2)*(y**2))
    return z

def V1(b):
    eps = 2; x = math.exp(eps/2) 
    z = np.zeros(len(b))
    for i in range(len(b)):
        y = math.exp(eps/b[i])
        z[i] = ((x**4)*(4*y+1) + 6*(x**2)*(y**2) + y**4 + 4*(y**3))/(3*((x**2-1)**2)*(y**2))
    return z
```

```python
b = np.linspace(1.6, 2)
fig = plt.figure()
plt.plot(b, V0(b))
plt.xlabel("$b$", fontsize = 18)
plt.ylabel("$V_{GPM}(0)$", fontsize = 18)
fig.savefig("BestVariance_GPM.png")
fig.savefig("figs/BestVariance_GPM.eps", bbox_inches="tight", pad_inches=0.05)
```

```python
b = np.linspace(1.6, 2)
fig = plt.figure()
plt.plot(b, V1(b))
plt.xlabel("$b$", fontsize = 18)
plt.ylabel("$V_{GPM}(1) = V_{GPM}(-1)$", fontsize = 18)
fig.savefig("WorstVariance_GPM.png")
fig.savefig("figs/WorstVariance_GPM.eps", bbox_inches="tight", pad_inches=0.05)
```

```python
def V0(b):
    eps = 2; x = math.exp(eps/2); y = math.exp(eps/b)
    return ((y+1)*(x**4 + 3*(x**2)*y + 3*(y**2) + y**3))/(3*((x**2-1)**2)*(y**2))

def V1(b):
    eps = 2; x = math.exp(eps/2); y = math.exp(eps/b)
    return ((x**4)*(4*y+1) + 6*(x**2)*(y**2) + y**4 + 4*(y**3))/(3*((x**2-1)**2)*(y**2))
```

```python
fig = plt.figure()
#plt.ylim(0.6, 1.25)
plt.xlim(-0.05, 1.05)
plt.plot([0,1], [V0(2),V1(2)], color='black', linestyle='dashed', label='$b = 2$')
plt.plot([0,1], [V0(1.8),V1(1.8)], color='blue', linestyle='dotted', label='$b = 1.8$')
plt.plot([0,1], [V0(1.7),V1(1.7)], color='green', linestyle='dashdot', label='$b = 1.7$')
plt.plot([0,1], [V0(1.6),V1(1.6)], color='red', linestyle='-', label='$b = 1.6$')
plt.legend(bbox_to_anchor=(0.49, 1.02), loc='lower center', borderaxespad=0, ncol=4, fontsize=11.5)
plt.xlabel("$t^2$", fontsize = 18)
plt.ylabel("$V_{GPM}(t)$", fontsize = 18)
fig.savefig("Variance_GPM.png")
fig.savefig("figs/Variance_GPM.eps", bbox_inches="tight", pad_inches=0.05)
```
