```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def V0(b):
    eps = 2; x = math.exp(eps/2); y = math.exp(eps/b)
    return ((y+1)*(x**4 + 3*(x**2)*y + 3*(y**2) + y**3))/(3*((x**2-1)**2)*(y**2))

def V1(b):
    eps = 2; x = math.exp(eps/2); y = math.exp(eps/b)
    return ((x**4)*(4*y+1) + 6*(x**2)*(y**2) + y**4 + 4*(y**3))/(3*((x**2-1)**2)*(y**2))
```

```python
fig = plt.figure()
#plt.ylim(0.6, 1.25)
plt.xlim(-0.05, 1.05)
eps = 2; x = math.exp(eps/2)
plt.plot([0,1], [2/(math.log(x)**2), 2/(math.log(x)**2)], color='m', linestyle='dashdot', label='LM')
plt.plot([0,1], [V0(2),V1(2)], color='black', linestyle='dashed', label='GPM ($b = 2$)')
plt.plot([0,1], [V0(1.6),V1(1.6)], color='red', linestyle='-', label='GPM ($b = 1.6$)')
plt.legend(bbox_to_anchor=(0.49, 1.02), loc='lower center', borderaxespad=0, ncol=4, fontsize=11.5)
plt.xlabel("$t^2$", fontsize = 18)
plt.ylabel("$V(t)$", fontsize = 18)
fig.savefig("GPM_LM.png")
fig.savefig("figs/GPM_LM.eps")
```
